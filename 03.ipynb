```python
sample = '''467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598..'''
```

```python
lines = sample.split('\n')
lines = [line.strip() for line in open('input/03.txt')]

n_col = len(lines[0])

lines = ['.' * n_col] + lines + ['.' * n_col]
lines = ['.' + line + '.' for line in lines]

n_row = len(lines)
n_col = len(lines[0])

is_symbol = lambda x: x not in '0123456789.'

def is_adjacent_to_symbol(x, y):
    is_adjacent = False
    is_adjacent = is_adjacent or is_symbol(lines[y-1][x-1]) or is_symbol(lines[y-1][x])  or is_symbol(lines[y-1][x+1])
    is_adjacent = is_adjacent or is_symbol(lines[y][x-1]) or is_symbol(lines[y][x+1])
    is_adjacent = is_adjacent or is_symbol(lines[y+1][x-1]) or is_symbol(lines[y+1][x])  or is_symbol(lines[y+1][x+1])

    return is_adjacent


total = 0

number = ''
is_adjacent = False
indicies = []
index_to_number = {}

for y in range(1, n_row-1):
    for x in range(1, n_col): # runs for last col to flush is_adjacent and number
        value = lines[y][x]
        if value.isdigit():
            number += value
            is_adjacent = is_adjacent or is_adjacent_to_symbol(x, y)
            indicies.append((x, y))
        elif number != '':
            print(number, is_adjacent)
            if is_adjacent:
                total += int(number)

                for index in indicies:
                    index_to_number[index] = int(number)
            
            number = ''
            is_adjacent = False
            indicies = []

total
```

```python
total = 0

for y in range(1, n_row-1):
    for x in range(1, n_col-1):
        if lines[y][x] == '*':
            nw = index_to_number.get((x-1, y-1), None)
            nn = index_to_number.get((x+0, y-1), None)
            ne = index_to_number.get((x+1, y-1), None)

            ww = index_to_number.get((x-1, y+0), None)
            ee = index_to_number.get((x+1, y+0), None)

            sw = index_to_number.get((x-1, y+1), None)
            ss = index_to_number.get((x+0, y+1), None)
            se = index_to_number.get((x+1, y+1), None)

            ratios = []
            n = 0

            if nn:
                ratios += [nn]
            else:
                ratios += [nw, ne]

            ratios += [ww, ee]

            if ss:
                ratios += [ss]
            else:
                ratios += [sw, se]

            ratios = [ratio for ratio in ratios if ratio is not None]

            total += (ratios[0] * ratios[1]) if len(ratios) == 2 else 0

total
            
```
